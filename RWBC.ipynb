```python
import requests 
import pandas as pd
import time
import re
import time
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

options = Options()
options.add_argument("--headless")
```

```python
from bs4 import BeautifulSoup as bs
```

```python
start = time.time()
```

```python
url= "https://reesesbookclub.com/picks" 
page = requests.get(url) 
soup = bs(page.content, 'html.parser') 
```

```python
urls = []
months_string = []
for book in soup.find_all('a', {'class':'chakra-link css-146zwt9'}):    
    url = 'https://reesesbookclub.com' + book.get('href')
    urls.append(url)    
for month in soup.find_all('span', {"class": re.compile("^css-1j")}):
    month = month.get_text()    
    months_string.append(month)    
```

```python
books = pd.DataFrame (urls, columns = ['urls'])
books['months'] = months_string
books['months'] = books['months'].str.replace('\'', ' ').str.replace('PICK','').str.replace('YA','')
books['title'] = books['urls'].str.split('#', expand=True)[1].str.replace('-', ' ').str.title()
```

```python
books
```

```python
all_genres = []
authors_list = []
for genre in urls[0:len(urls)]:
    driver = webdriver.Chrome('C:/Users/marti/Sonka/python/chromedriver.exe')
    all_genres.append(genre.split('picks/')[1])     
    driver.get(genre)
    author = driver.find_elements_by_xpath('//p[@class="chakra-text css-kotshv"]')[0].text
    authors_list.append(author)  
    
```

```python
books['authors'] = authors_list
```

```python
end = time.time()
print(end - start) #10 mins
```

```python
books.to_csv('books_rwbc.csv', index = False)
```
